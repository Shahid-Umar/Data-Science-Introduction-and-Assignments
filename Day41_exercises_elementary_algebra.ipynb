## **Basic Operations and Simplifications**

1. Simplify: 7x + 3x
   - `11x`
2. Solve for x: 5x – 2 = 13
   - `5x = 13+2 ==>> 5x = 15 ==>> x = 15/5`
   - `x = 3`
3. Simplify: 4(y + 3)
   - `4y + 12`
4. Simplify the expression: (2x^2 + 6x) / 2x
   - `2x*x/2x + 6x/2x`
   - `x = 3`
5. Solve for y: 3y + 4y – 5 = 10
   - `7y = 10+5 ==>> 7y = 15`
   - `y = 15/7`
6. Simplify: 8z – 3z + 2
   - `5z + 2`

## **Equations and Inequalities**

7. Solve for x: 2x + 7 = 17
   - `2x = 17 - 7 ==>> 2x = 10 ==>> x = 10/2`
   - `x = 5`
8. Solve the inequality: 4x – 3 > 9
   - `4x > 9+3 ==>> 4x > 12 ==>> x > 12/4`
   - `x > 3`
9. Solve for x: 3x – 5 = 10
    - `3x = 10+5 ==>> 3x = 15 ==>> x = 15/3`
    - `x = 5`
10. Solve the inequality: 5y + 6 <= 16
    - `5y <= 16-6 ==>> 5y <= 10 ==>> y <= 10/5`
    - `y <= 2`

## **Functions and Graphs**

11. Find f(2) if f(x) = 2x + 5
    - `f(2) = 2*2 + 5 = 4+5 = 9`
12. For g(x) = x^2 – 4, find g(3)
    - `g(3) = 3^2 - 4 = 9-4 = 5`
13. Determine the slope of the line represented by y = 3x + 2
    - `We know that the slop of line is y=mx+2 here m = slop = 3`
    - ![Slop](slop.png)
14. Plot the function f(x) = x + 3 and identify the y-intercept
    - `we will get pairs (-3, 0) (-2, 1) (-1, 2) (0, 3) (1, 4) (2, 5) when putting the value of x = -3,-2,-1,0,1,2`
    - ![y-intercept Plot](y-intercept.jpg)
15. Determine the x-intercept for the function y = -2x + 6
    - `For x-intercept, we can suppose y = 0`
    - `0 = -2x + 6 ==>> -2x = -6 ==>> x = -6/-2 ==>> x = 3`
    - `So, x-intercept point is (3,0)`

## **Mixed and Challenge Questions**

21. If f(x) = 2x^2 + 3x – 5, find f(2)
    - `f(2) =  2*2^2 + 3*2 - 5 = 8 + 6 - 5 = 14 - 5`
    - `f(2) = 9`
22. Solve for x: (x + 3) / 2 = 7
    - `x + 3 = 7*2 ==>> x = 14 - 3`
    - `x = 11`
23. Simplify and solve: 2(x – 3) + 5 = 11
    - `2x - 6 + 5 = 11 ==>> 2x - 1 = 11 ==>> 2x = 11+1 ==>> x = 12/6`
    - `x = 2`
24. If g(x) = x^3 – 2x + 1, calculate g(-1)
    - `g(-1) = -1^3 - 2(-1) + 1 = -1 + 2 + 1 = -1 + 3`
    - `g(-1) = 2`
25. Solve the inequality: -3x + 4 >= 13
    - `-3x >= 13-4 ==>> x >= 9/(-3)`
    - `x >= -3`

## **Advanced Applications**

26. For the function h(x) = sqrt(x + 6), find h(4)
    - `h(4) = sqrt(4 + 6) = sqrt(10)`
    - `h(4) = 100 Note: sqrt(10) means 10^2=100`
27. Solve for x: 1/2x – 1/4 = 3/4x + 1
    - `1/2x - 3/4x = 1 + 1/4 ==>> (2 - 3)/4x = (4 + 1)/4`
    - `-2/4x = 5/4 ==>> -1/2x = 5/4 ==>> x = (5/4) * (-2)`
    - `x = -10/4 ==>> x = -5/2`
28. If f(x) = 1/x, find the y-intercept of the graph
    - `If y-intercept required, we substitute x = 0`
    - `f(0) = 1/0 Here value is undefined because of zero division`
    - `So, the above function does not have y-intercept`
29. Solve the system of equations: x^2 + y^2 = 25, x + y = 9
    - `For this solution, we will add both equations`
    - `x^2 + y^2 + x + y = 25 + 9`
    - `After addition we get x^2 + y^2 x + y = 34`
30. Given the function f(x) = 3^x, find f(2) and f(-1)
    - `f(2) = 3^2 = 9`
    - `f(-1) = 3^(-1) = 3`
