   - `The mode is 6, because mode is a maximum occurrence in data`
4. Calculate the range of the following numbers: 5, 8, 12, 15, 22.
   - `We know that Range = Largest Number - Smallest Number`
   - `So, the Range = 22-5 = 17`
5. If the first quartile (Q1) is 5 and the third quartile (Q3) is 15, what is the interquartile range?
   - `The interquartile (IQR) = Q3 - Q1 = 15-5 = 10`

7. What is the probability of rolling a number greater than 4 on a standard six-sided die?
   - `We know that total outcomes in die is 6, so more than 4 number outcomes is only 2 (i,e 5 or 6)`
   - `Probability = Favorable outcomes / Total outcomes = 2/6 = 1/3 = 0.333`
   - `The probability for occurrence of each number greater than 4 in die is 33.33%`
8. If a coin is flipped three times, what is the probability of getting exactly two heads?
   - `Flipping coin are two possible outcomes (i,e H or T)`
   - `When flipping a coin three times then possible outcomes = 2^3 = 8 (HHH, HHT, HTH, HTT, THH, THT, TTH, TTT)`
   - `the exactly two head outcomes from above possible outcomes are HHT, HTH, and THH (3 outcomes)`
   - `Probability = possible outcomes / total outcomes = 3/8 = 0.375`
   - `So, the probability for occurrence of exactly two head in three times filliping is 37.5%`

## Chapter 8: Pre-Algebra Problem Solving

# Day38 Assignment Exercises (8/11)
> Click for blog reading *[The Ultimate Zero-to-Hero Guide to Pre-Algebra](https://codanics.com/the-ultimate-zero-to-hero-guide-to-pre-algebra/)*
> Chapter-8 : Pre-Algebra Problem Solving

1. A recipe calls for 3 cups of flour for every 4 cups of sugar. How much sugar is needed if 6 cups of flour are used?
   - `In the above statement ratio is '3 cups flour : 4 cups sugar' = 3:4`
   - `We can setup a proportion like 3 cups flour / 4 cups sugar = 6 cups flour / x cups sugar = 3/4 = 6/x`
   - `By cross multiplication we solve it as 3x = 24`
   - `x = 24/3 = 8`
   - `So, we can say, if 6 cups flour are used then 8 cups sugar is need that maintain the 3:4 ratio.`
2. An art piece is sold for $120, which is a 20% profit over the cost. What was the cost price?
   - `Suppose original cost of art is 'C'`
   - `Selling cost = original cost + 20% of original cost = C + 20% of C`
   - `120 = C + 0.2C ==> 120 = 1.2C ==> C = 120/1.2 = 100`
   - `So, the original cost of art is $100`
3. A rectangular garden has a length of 20 meters and a width that is half of its length. What is the area of the garden?
   - `The area of rectangle is (A) = Length * Width`
   - `The area of garden is length (20 meters) * width (length/2) = 20 * (20/2) = 20 * 10 = 200`
   - `So, the area of garden is 200 square meters`
4. If 3 pencils cost 45 cents, how much would 10 pencils cost?
   - `The cost of pencil is 'C'`
   - `Using set proportion 3 pencils / 45 cents cost = 10 pencils / C ==> 3/45 = 10/C`
   - `By cross-multiplication method 3C = 10*45 ==> 3C = 450 ==> C = 450/3 ==> C = 150`
   - `So, the cost of 10 pencils is 150 cents`
5. You have a rope 30 meters long, and you need to cut it into pieces that are 2.5 meters each. How many pieces can you cut?
   - `Rope Total length (L) = 30, Single Piece Length (S) = 2.5, No of Rope Pieces = Rp`
   - `Now we have expression 2.5 * Rp = 30 ==> Rp = 30/2.5 = 12`
   - `So, we drive that 12 pieces of ropes we can get.`
6. Mike read twice as many books this year as last year. If he read 5 books last year, how many books did he read this year?
   - `Suppose we consider this year read is x`
   - `As per above statement x = 2 * last year read`
   - `x = 2 * 4 = 8 books read this year`
7. If a car travels 150 kilometers in 2 hours, what is its average speed?
   - `Average speed of car (M) = Total Distance / Total Time`
   - `M = 150/2 = 75 kilometer`
8. Solve for x in the equation 5x − 6 = 2x + 9.
   - `5x - 2x = 9 + 6 ==> 3x = 15 ==> x = 15/3`
   - `So, x = 5`
9. A box contains red and blue marbles. If the ratio of red to blue marbles is 3:4 and there are 12 red marbles, how many blue marbles are there?
    - `The above statement can be solved by proportion method 3/4 = 12/x`
    - `By cross multiplication ==> 3x = 48 ==> x = 48/3 = 16`
    - `16 blue marbles will be there to complete the ratio of 3:4`
10. If the price of a jacket after tax is $108 and the tax rate is 8%, what is the original price of the jacket before tax?
    -  `Suppose original cost of jacket is 'P'`
    -  `Price after tax = original cost + 8% of original cost = P + 8% of P`
    -  `108 = P + 0.08P ==> 108 = 1.08C ==> C = 108/1.08 = 100`
    -  `So, the original price of jacket is $100`
11. Convert a speed of 60 miles per hour to feet per second.
    - `we know that 1 mile is equal to 5280 feet, and 1 hour is equal to 3600 seconds.`
    - `60 miles per hour * (5280 feet per mile) / (3600 seconds per hour) = (60 * 5280)/3600 = 88`
    - `So, a speed of 60 miles per hour is equivalent to 88 feet per second.`
12. How many different outfits can you make with 4 shirts and 3 pairs of pants?
    - `Total outfits = Number of shirts * Number of pants = 3 * 4 = 12`
    - `So, we can make 12 outfits`
13. Calculate the surface area of a cylinder with a radius of 3 units and a height of 5 units.
    - `The area of surface (A) = (2 * π * r * h) + (2 * π * r^2) Here we have r=3 and h=5`
    - `A = (2 * 3.14 * 3 *5) + (2 * 3.14 * 3^2) = 94.2 + 56.52 = 150.72`
    - `So, the surface area of the cylinder 150.72 square units.`
14. A bookstore offers a 15% discount on a book that originally costs $35. What is the discount amount?
    - `Discounted amount = Original cost of book * 15% discount = (35 * 15)/100 = 525/100 = 5.25`
    - `So, the discounted amount is $5.25`
15. If you have a square plot of land with an area of 49 square meters, what is the perimeter of the plot?
    - `We know that the perimeter of square (P) = 4 √Area of square Here '√' the root of square`
    - `P = 4 √49 = 4 * 7 = 28`
    - `So, we can say the perimeter of plot is 28 square units`
