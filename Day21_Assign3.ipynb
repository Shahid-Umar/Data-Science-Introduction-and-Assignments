# 9. Spider Chart

```python
# All columns are:
#'total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size'

# Sample data
data = {
    'Tip_Category': ['Total Bill', 'Tip', 'Size'],
    'Total Bill': [0, 0, 0],
    'Tip': [1, 1, 1],
    'Size': [1, 1, 1]
}

# Create a pandas dataframe from the dictionary
df2 = pd.DataFrame(data)

# Number of variables
categories = list(df2['Tip_Category'])
N = len(categories)

# Set the angle for each metric's axis
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

# Initialize the spider plot
plt.figure(figsize=(8, 5))
ax = plt.subplot(111, polar=True)

# Plotting for Total Bill
values = df2['Total Bill'].tolist()
values += values[:1]  # Repeat the first value to close the circular graph
ax.plot(angles, values, linewidth=2, linestyle='solid', label='Total Bill1')
ax.fill(angles, values, 'b', alpha=0.1)

# Plotting for Tip
values = df2['Tip'].tolist()
values += values[:1]  # Repeat the first value to close the circular graph
ax.plot(angles, values, linewidth=2, linestyle='solid', label='Tip')
ax.fill(angles, values, 'r', alpha=0.1)

# Plotting for Size
values = df2['Size'].tolist()
values += values[:1]  # Repeat the first value to close the circular graph
ax.plot(angles, values, linewidth=2, linestyle='solid', label='Size')
ax.fill(angles, values, 'r', alpha=0.1)

# Add the axis labels
plt.xticks(angles[:-1], categories, color='grey', size=12)
ax.set_rlabel_position(30)
plt.yticks([1,2,3], ["1","2","3"], color="grey", size=8)
plt.ylim(0, 5)

# Add title and legend
plt.title('Comparison between Total_Bill, Tip and Size')
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Save the plot to a file
plt.savefig('./assign_plots/spiderchart_D21A3_01.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

```python
# All columns are:
#'total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size'

# Sample data
data = {
    'Total Bill': [43.11, 40.17, 50.81, 48.17],
    'Tip': [6.70, 4.73, 10.00, 6.50],
    'Size': [90, 30, 120, 60],  # Determines the size of bubbles
    'Labels': ['Thursday', 'Friday', 'Saturday', 'Sunday']
}

# Creating the bubble chart
fig = go.Figure(data=go.Scatter(
    x=data['Total Bill'],
    y=data['Tip'],
    mode='markers',
    text=data['Labels'],
    marker=dict(
        size=data['Size'],
        sizemode='diameter',  # 'diameter' ensures that the size values correspond to the diameter of the bubbles
        opacity=0.6,
        color=data['Tip'],  # Coloring the bubbles based on Y values
        colorscale='Viridis',
        colorbar=dict(title='Colorbar')
    )
))

# Setting the layout for the plot
fig.update_layout(
    title='Bubble Chart using Plotly',
    title_x=0.5, # To align title in center
    xaxis_title='Total Bill Daily Averages',
    yaxis_title='Daily Tip Averages',
    showlegend=False
)

# Display the plot
fig.show()

# Save the plot to an HTML file
fig.write_html("./assign_plots/bubblechart_D21A3_01.html")

# save png file with 300 dpi
fig.write_image("./assign_plots/bubblechart_D21A3_01.png", scale=3)
```
